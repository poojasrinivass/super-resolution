### Deep Network Cascade for image Super-resolution 

#### NLSS implementation 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img_gray = cv2.imread('../image/sky.jpeg',0)
```

```python
def gaussiand_2_level_pyramid(img):
    
    blur_1 = cv2.GaussianBlur(img,(3,3),0)
    [h,w]  = blur_1.shape
    ## down_sample the blur1 and apply gaussian again 
    d_blur_1 = cv2.resize(blur_1, (0,0), fx=0.5, fy=0.5) 
    d_blur_2 = cv2.GaussianBlur(d_blur_1,(3,3),0)
    ## upsample image 
    blur_2 = cv2.resize(d_blur_2,(h,w),0,0, interpolation = cv2.INTER_NEAREST)
    return blur_1, blur_2
```

```python
def NLSS_serach(img_gray):
    ## apply bicubic interpolation 
    img_gray_scale = cv2.resize(img_gray,(0,0),fx=1.2,fy=1.2, interpolation = cv2.INTER_CUBIC)
    
    [h,w] = img_gray_scale.shape
    out_img = np.zeros([h,w],dtype='float')
    
    for row in range (0,h-7,7):
        for col in range (0,w-7,7):
            
            patch = img_gray_scale [row:row+7,col:col+7]
            ## apply blur in pathand perform down sampling 
            blur_1, blur_2 = gaussiand_2_level_pyramid(patch)
            mean_patch = blur_1/2.0 + blur_2/2.0        ## averaging instead of gaussina
            out_img[row:row+7,col:col+7] = mean_patch
    return out_img
```

```python
blur_1, blur_2 = gaussiand_2_level_pyramid(img_gray)
```

```python
NLSS_img = NLSS_serach(img_gray)
```

```python
fig=plt.figure(figsize=(16, 16))
columns = 1
rows = 2
fig.add_subplot(rows, columns, 1)
plt.axis("off")
plt.title("gray_image")
plt.imshow(img_gray,cmap='gray')
fig.add_subplot(rows, columns, 2)
plt.axis("off")
plt.title("NLSS_image")
plt.imshow(NLSS_img,cmap='gray')
```

## Autoencoder basic module 

```python
import keras
from keras.layers import Input,Conv2D,MaxPooling2D,UpSampling2D
from keras.models import Model
from keras.optimizers import RMSprop
```

```python
## there might be parameter changes needed basic auto encoder structre

def autoencoder(img):
    #encoder
    #input = 8 x 8 x 1 (wide and thin)
    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(img) #8 x 8 x 32
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1) #4 x 4 x 32
    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1) #4 x 4 x 64
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2) #2 x 2 x 64
    conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(pool2) #2 x 2 x 128 (small and thick)

    #decoder
    conv4 = Conv2D(128, (3, 3), activation='relu', padding='same')(conv3) #2 x 2 x 128
    up1 = UpSampling2D((2,2))(conv4) # 14 x 14 x 128
    conv5 = Conv2D(64, (3, 3), activation='relu', padding='same')(up1) # 4 x 4 x 64
    up2 = UpSampling2D((2,2))(conv5) # 28 x 28 x 64
    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(up2) # 8 x 8 x 1
    return decoded
```
